OVERLOADING RELATIONAL OPERATOR :-
--------------------------------------


```python
class student:
    def __init__(self,name,marks):
        self.name=name
        self.marks=marks
    def __gt__(self,other):
        return self.marks>other.marks
    def __le__(self,other):
        return self.marks<other.marks
ob1=student("abc",100)
ob2=student("def",200)
print(ob2>ob1)
print(ob1>ob2)
```

METHOD OVERLOADING:-
IN PYTHON OVERLOADING IS NOT POSSIBLE.

CONSTRUCTOR OVERLOADING.
CANT OVERLOADING BUT OVERRIDING IS POSSIBLE.

```python
class mycls:
    def __init__(self):
        print("hii")
    def __init__(self,a):
        print("bye")
    def __init__(self,b,c):
        print("hello")
obj=mycls(10,20)
```

SUPER METHOD:---
 

```python
class class1:
    def myfunction(self):
        print("pritam have a 1 kg gold")
    def marry(self):
        print("marry me")
class class2(class1):
    def marry(self):
        super().marry()
        print("saadi mubarak")

obj=class2()
obj.myfunction()
obj.marry()
```
