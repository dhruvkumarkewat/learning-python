```python
my_dict = dict()

while True:
    print("\nMain Menu:")
    print("1. Create")
    print("2. Read")
    print("3. Update")
    print("4. Delete")
    print("5. Exit")
    
    try:
        ch = int(input("Enter your choice: "))
    except ValueError:
        print("Please enter a valid number.")
        continue

    if ch == 1:
        print("\n--- CREATE DATA IN DICTIONARY ---")
        n = int(input("Enter the number of entries to add: "))
        for i in range(n):
            key = input(f"Enter key {i + 1}: ")
            value = input(f"Enter value for key '{key}': ")
            my_dict[key] = value
        print("Data added successfully.")

    elif ch == 2:
        print("\n--- READ DICTIONARY DATA ---")
        while True:
            print("\n1. View all data")
            print("2. View specific data by key")
            print("3. Go back to main menu")
            
            try:
                read_ch = int(input("Enter your choice: "))
            except ValueError:
                print("Please enter a valid number.")
                continue

            if read_ch == 1:
                print("Full Dictionary:")
                print(my_dict)

            elif read_ch == 2:
                key = input("Enter the key to search: ")
                if key in my_dict:
                    print(f"Value for key '{key}': {my_dict[key]}")
                else:
                    print("Key not found.")

            elif read_ch == 3:
                print("Returning to main menu.")
                break
            else:
                print("Invalid choice. Please try again.")

    elif ch == 3:
        print("\n--- UPDATE DICTIONARY DATA ---")
        key = input("Enter the key to update: ")
        if key in my_dict:
            new_value = input(f"Enter new value for key '{key}': ")
            my_dict[key] = new_value
            print("Data updated successfully.")
        else:
            print("Key not found.")

    elif ch == 4:
        print("\n--- DELETE DATA FROM DICTIONARY ---")
        key = input("Enter the key to delete: ")
        if key in my_dict:
            del my_dict[key]
            print("Data deleted successfully.")
        else:
            print("Key not found.")

    elif ch == 5:
        print("Exiting the program.")
        break

    else:
        print("Invalid choice. Please enter a number between 1 and 5.")
```
