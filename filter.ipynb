FILTER FUNCTION:-
----------------------------
WE CAN USE FILTER FUNCTION TO FILTER VALUES FROM THE GIVEN SEQUENCE WE CAN USE THE FILTER FUNCTION TO FILTER FROM THE GIVEN SEQUENCE BASED ON SOME CONDITION .

```python
def even(x):
    if (x%2==0):
        return True
    else:
        return False
l=[0,2,5,10,15,20,25]
l1=list(filter(even,l))   #in filter use function and sequence 
print(l1)
```

```python
l=[0,5,10,15,20,25,30]
l1=list(filter(lambda x:x%2==0,l))
print(l1)
l2=list(filter(lambda x:x%2!=0,l))
print(l2)
```

map():- for every element present in given sequence , apply some functionality new element with required modification.For this requirement we should go for map() function.

```python
def myfunc(x):
    return 2*x
l=[1,2,3,4,5]
l1=list(map(myfunc,l))
print(l1)
```

Reduce:-
------------------
reduce function reduces the sequence of an element into an single element.
It is present in module which is functools

```python
from functools import *
l=[10,20,30,40,50]
result=reduce(lambda x,y:x+y,l)
print(result)
```

FUNCTION DECORATORS:- IMP
-------------------------------------
Decorators is a function which can take function as an arguements and extend its functionality and returns modified function with extended functionality.

```python
def myfunc(name):
    print(f"hello my name is {name}")

def decor(func):
    def inner(name):
        if name=="gaurav" :
            print("hello gaurav")
        else:
            print("sorry gaurav")
    return inner
@decor
def myfunc(name):
    print(f"hello my name is {name}")
myfunc("gaurav")
myfunc("dhruv")
```

TYPES OF VARAIBLE:-
-------------------------------------------
In side python there will be three types of varaible allowed.
---------------------------------------------------------------
1.Instance varaible (object level varaible)
inside the constructor, inside method,outside the class.

2.Static varaible(class level varaible)
3.Local varaible(method level varaible)



```python
class test:                  #outside the class
    def __init__(self):
        self.a=10
        self.b=20
    def method1(self):
        self.c=30
obj=test()
obj.method1()
obj.d=40
print(obj.__dict__)
print(obj.a)
```

how to delete instance varaible from access

```python
class test:                  
    def __init__(self):
        self.a=10
        self.b=20
        self.c=30
        self.d=40
    def method1(self):
        del self.d
obj=test()
print(obj.__dict__)
obj.method1()
print(obj.__dict__)
```

```python
class test:                  
    def __init__(self):
        self.a=10
        self.b=20
        self.c=30
        self.d=40
    def method1(self):
        del self.d
obj=test()
print(obj.__dict__)
obj.method1()
print(obj.__dict__)
del obj.c
print(obj.__dict__)
```

STATIC VARAIBLE:-
-------------------
IF the value of varaible is not valid from object to object such type of varaible we have to declare inside the class directly
but outside of methods , such type of varaible is called statsic varaible.

```python
class Test:
    x=10
    def __init__(self):
        self.y=20
t1=Test()
t2=Test()
print(t1.x,t1.y)
Test.x=898
print(t1.x)
print(t1.y)
Test.y=567
print(t2.y)

```

PASSING MEMBER OF ONE CCLASS TO ANOTHER CLASS:-
---------------------------------------------------


```python
class Employee:
    def _init_(self,eno,ename,esal):
        self.eno=eno
        self.ename=ename
        self.esal=esal
    def display(self):
        
        print("enumber=",self.eno)
        print("enumber=",self.esal)
class mycls:
    def modify(emp):
        emp.esal=emp.esal+1000
        emp.display()
obj=Employee(100,"gaurav",1000)
mycls.modify(obj)
```
