```python
my_tuple = tuple()

while (True):
    print("\n--- MENU ---")
    print("Press 1 for CREATE")
    print("Press 2 for READ")
    print("Press 3 for UPDATE")
    print("Press 4 for DELETE")
    print("Press 5 for EXIT")

    ch = int(input("Enter your choice: "))

    if ch == 1:
        print("\nCREATE DATA IN TUPLE")
        n = int(input("Enter the number of data: "))
        temp_list = list(my_tuple)   # convert tuple → list
        for i in range(n):
            data = input(f"Enter data {i+1}: ")
            temp_list.append(data)
        my_tuple = tuple(temp_list)  # convert list → tuple

    elif ch == 2:
        print("\nREAD TUPLE DATA")
        while True:
            print("\n1. Show all data")
            print("2. Show specific data")
            print("3. Exit read mode")
            choice = int(input("Enter choice: "))

            if choice == 1:
                print(my_tuple)
            elif choice == 2:
                n = input("Enter data to search: ")
                if n in my_tuple:
                    print(f"Data found: {n}")
                else:
                    print("Data not found")
            elif choice == 3:
                print("Exiting read mode")
                break
            else:
                print("Invalid choice!")

    elif ch == 3:
        print("\nUPDATE DATA IN TUPLE")
        data = input("Enter data to update: ")
        if data in my_tuple:
            temp_list = list(my_tuple)
            ind = temp_list.index(data)
            new_data = input("Enter new value: ")
            temp_list[ind] = new_data
            my_tuple = tuple(temp_list)
            print("Data updated successfully!")
        else:
            print("Data not found in tuple")

    elif ch == 4:
        print("\nDELETE DATA FROM TUPLE")
        data = input("Enter data to delete: ")
        if data in my_tuple:
            temp_list = list(my_tuple)
            temp_list.remove(data)
            my_tuple = tuple(temp_list)
            print("Data deleted successfully!")
        else:
            print("Data not found in tuple")

    elif ch == 5:
        print("EXITING PROGRAM...")
        break

    else:
        print("Invalid choice! Try again.")
```

BITWISE OPERATOR IN PYTHON:-
--------------------------

1. BITWISE AND OPERATOR(&)
IF BOTH BITS ARE 1 THEN RESULT SHOULD BE ONE.
2. BITWISE OR OPERATOR(|)
IF BOTH BITS ARE ZERO THAN RESULT SHOULD BE ZERO.
3. XOR OPERATOR(^)
IF BITS ARE DIFFERENT THEN VALUE IS ONE OTHERWISE ZERO,
4. NOT OPERATOR(~)
5. Left shift operator(<<)

```python
a=12
b=13
c=a&b
print(c)
n=int(input("Enter the number:"))
x=n&1
if(x==0):
    print("even")
else:
    print("odd")
```
