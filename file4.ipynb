write a program to show the markssheeet of any student of three subject input take by user by following codition marks should be 0 to 100 , if user fail in one sub then it will be fail but also show their real world result and less than 45 grade c , more than 45 b

```python
# Marksheet program for 3 subjects

# Taking input
marks = []
for i in range(1, 4):
    mark = int(input(f"Enter marks for subject {i} (0-100): "))
    
    # validation
    if mark < 0 or mark > 100:
        print(" Invalid input! Marks should be between 0 and 100.")
        exit()
    
    marks.append(mark)

# Check fail condition
if any(m < 33 for m in marks):
    print("\nResult:  FAIL")
    print("Reason: You failed in one or more subjects.")
else:
    # Calculate percentage
    total = sum(marks)
    percentage = total / 3
    
    # Grade system
    if percentage < 45:
        grade = "C"
    elif percentage <= 60:
        grade = "B"
    else:
        grade = "A"
    
    print("\nResult:  PASS")
    print(f"Total Marks: {total}/300")
    print(f"Grade: {grade}")
```

```python
# Marksheet program for 3 subjects

# Taking input
marks = []
for i in range(1, 4):
    mark = int(input(f"Enter marks for subject {i} (0-100): "))
    
    # validation
    if mark < 0 or mark > 100:
        print("❌ Invalid input! Marks should be between 0 and 100.")
        exit()
    
    marks.append(mark)

# Check fail condition
if any(m < 33 for m in marks):
    print("\nResult: ❌ FAIL")
    print("Reason: You failed in one or more subjects.")
else:
    # Calculate percentage
    total = sum(marks)
    percentage = total / 3
    
    # Grade system
    if percentage < 45:
        grade = "C"
    elif percentage <= 60:
        grade = "B"
    else:
        grade = "A"
    
    print("\nResult: ✅ PASS")
    print(f"Total Marks: {total}/300")
    print(f"Percentage: {percentage:.2f}%")
    print(f"Grade: {grade}")
```
