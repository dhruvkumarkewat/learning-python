```python
# Simple Login & Account Balance Management System

# Global data
stored_user = "admin"
stored_pass = "1234"
balance = 1000.0  # initial balance

# Function to handle login
def login():
    user = input("Enter Username: ")
    password = input("Enter Password: ")

    if user == stored_user and password == stored_pass:
        print("\nLogin Successful!\n")
        return True
    else:
        print("\nInvalid username or password.\n")
        return False

# Function to show menu
def show_menu():
    print("\n===== Main Menu =====")
    print("1. Check Balance")
    print("2. Deposit Money")
    print("3. Withdraw Money")
    print("4. Logout & Exit")

# Function to check balance
def check_balance():
    print(f"\nYour current balance is: Rs. {balance}")

# Function to deposit
def deposit():
    global balance
    try:
        amount = float(input("Enter amount to deposit: "))
        if amount > 0:
            balance += amount
            print(f"Deposited Rs. {amount} successfully.")
            check_balance()
        else:
            print("Invalid deposit amount.")
    except ValueError:
        print("Please enter a valid number.")

# Function to withdraw
def withdraw():
    global balance
    try:
        amount = float(input("Enter amount to withdraw: "))
        if amount > 0 and amount <= balance:
            balance -= amount
            print(f"Withdrawn Rs. {amount} successfully.")
            check_balance()
        elif amount > balance:
            print("Insufficient balance!")
        else:
            print("Invalid withdrawal amount.")
    except ValueError:
        print("Please enter a valid number.")

# Main program
print("===== Welcome to Login & Account Balance Management System =====")

if login():
    choice = 0
    while choice != 4:
        show_menu()
        try:
            choice = int(input("Enter your choice: "))
            if choice == 1:
                check_balance()
            elif choice == 2:
                deposit()
            elif choice == 3:
                withdraw()
            elif choice == 4:
                print("Logging out... Goodbye!")
            else:
                print("Invalid choice! Try again.")
        except ValueError:
            print("Please enter a valid number.")
else:
    print("Exiting program due to failed login.")
```

There will be a two list in which objects are [12,11,13,15,17,19]and [23,11,19,17,67,31]
find the number which have prime in both and sum them and which have greater than four factor
also show them in another list.

```python
# Lists
list1 = [12, 11, 13, 15, 17, 19]
list2 = [23, 11, 19, 17, 67, 31]

# Function to check prime
def is_prime(n):
    if n < 2:
        return False
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True

# Common elements
common = [x for x in list1 if x in list2]

# Prime numbers in common
prime_common = [x for x in common if is_prime(x)]

# Sum of prime numbers
prime_sum = sum(prime_common)

# Numbers with more than 4 factors
def factor_count(n):
    count = 0
    for i in range(1, n + 1):
        if n % i == 0:
            count += 1
    return count

greater_than_4_factors = [x for x in common if factor_count(x) > 4]

# Output
print("Common prime numbers:", prime_common)
print("Sum of common primes:", prime_sum)
print("Numbers with more than 4 factors:", greater_than_4_factors)
```
